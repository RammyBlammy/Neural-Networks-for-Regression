```python
import numpy as np
import pandas as pd
from keras.layers import Dense, Activation
from keras.models import Sequential
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
def get_model(n1, n2, activation, optimizer, countLayers, X_train, y_train, batch, epoch, verb):
  model = Sequential()
  model.add(Dense(units=n1,  input_dim=X_train.shape[1], kernel_initializer='normal', activation=activation))
  i = 0
  for i in range(countLayers):
    model.add(Dense(units=n2, kernel_initializer='normal', activation=activation))
  model.add(Dense(1, kernel_initializer='normal', activation='linear')) 
  model.compile(loss='mean_squared_error', optimizer=optimizer, metrics = ['mae'])
  model.fit(X_train, y_train , batch_size = batch, epochs = epoch, verbose=verb)
  return model
```

```python
# Importing the dataset
#dataset = np.genfromtxt("3D_spatial_network.txt", delimiter=',')
dataset = pd.read_csv('3D_spatial_network.txt', sep=",", nrows=3000, header = None)
dataset.columns = ['ID', 'LONGITUDE', 'LATITUDE', 'ALTITUDE'];

y = dataset['ALTITUDE'];
y = y.values.reshape(-1,1)
X = dataset.drop('ALTITUDE', axis = 1);
X
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
PredictorScaler=StandardScaler()
TargetScaler=StandardScaler()
 
# Storing the fit object for later reference
PredictorScalerFit=PredictorScaler.fit(X)
TargetVarScalerFit=TargetScaler.fit(y)
 
# Generating the standardized values of X and y
X=PredictorScalerFit.transform(X)
y=TargetVarScalerFit.transform(y)

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.03, random_state = 42)
```

```python
model = get_model(350, 300, "relu", "adam", 2, X_train, y_train, 5, 100, 1)
y_pred = model.predict(X_test).flatten()
```

```python
train_ev = model.evaluate(X_train, y_train, verbose = 0)
test_ev = model.evaluate(X_test, y_test, verbose = 0)
print(train_ev, test_ev)
```

```python
print ("вычисл. \t тестов.\t погрешность")
for i in range(50):
  print (y_pred[i],"\t" , y_test[i],"\t",abs(y_test[i] - y_pred[i]))
```

```python
plt.plot(y_test, color = 'red', label = 'Real data')
plt.plot(y_pred, color = 'blue', label = 'Predicted data')
plt.title('Prediction')
plt.legend()
plt.show()
```
